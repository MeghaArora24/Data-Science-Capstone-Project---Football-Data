```python
merged_df = pd.merge(games, game_lineups, on='game_id', how = 'left')
merged_df
```

```python
merged_df = pd.merge(merged_df, game_events, on='game_id', how = 'left')
merged_df
```

```python
merged_df = pd.merge(merged_df, appearances, on='game_id', how = 'left')
merged_df
```

```python
final_merged_df = pd.merge(merged_df, players, on='player_id', how = 'left')
final_merged_df
```
